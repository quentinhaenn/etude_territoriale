```python
df_loyer_appart = pd.read_csv("../data/loyer_appart.csv", encoding='latin-1', sep=";")
df_loyer_maison = pd.read_csv("../data/loyer_maison.csv", encoding='latin-1', sep=";")
df_niveau_vie = pd.read_excel("../data/NdV_med.xlsx", header=3, index_col=0, skipfooter=3)
df_salaire_tp = pd.read_excel("../data/SEQT.xlsx", header=3, index_col=0, skipfooter=3)
df_loyer_appart
```

```python
df_loyer_appart['loypredm2'] = df_loyer_appart['loypredm2'].apply(lambda x: float(x.replace(',','.')))
df_loyer_appart['R2_adj'] = df_loyer_appart['R2_adj'].apply(lambda x: float(x.replace(',','.')))
df_loyer_appart.dtypes
```

```python
df_loyer_maison['loypredm2'] = df_loyer_maison['loypredm2'].apply(lambda x: float(x.replace(',','.')))
df_loyer_maison['R2_adj'] = df_loyer_maison['R2_adj'].apply(lambda x: float(x.replace(',','.')))
df_loyer_maison.dtypes
```

```python
df_loy_app_by_dep = df_loyer_appart[['loypredm2','R2_adj','DEP']].groupby(by="DEP").median()
df_loy_app_by_dep.rename(columns={'loypredm2' : 'loyer_app_m2', 'R2_adj' : 'R2_app'},inplace=True)
df_loy_app_by_dep
```

```python
df_loy_maison_by_dep = df_loyer_maison[['loypredm2','R2_adj','DEP']].groupby(by="DEP").median()
df_loy_maison_by_dep.rename(columns={'loypredm2' : 'loyer_maison_m2', 'R2_adj' : 'R2_maison'},inplace=True)
df_loy_maison_by_dep
```

```python
df_consolidee = df_loy_app_by_dep.join(df_loy_maison_by_dep).join(df_niveau_vie['Niveau de vie annuel médian']).join(df_salaire_tp['Salaire net moyen mensuel en EQTP'])
df_consolidee
```

On considère un appartement de 52m2 pour une personne seule (données proposées par l'ANSIL étant à la base de leur étude), ou une maison de 80m2 pour deux personnes.

```python
df_consolidee["loyer_app_total"] = df_consolidee["loyer_app_m2"]*52
df_consolidee["loyer_maison_total"] = df_consolidee["loyer_maison_m2"]*52
df_consolidee["RAV_niv_vie_median_app"] = df_consolidee["Niveau de vie annuel médian"]-df_consolidee["loyer_app_total"]*12
df_consolidee["RAV_salaire_moyen_eqtp_app"] = df_consolidee["Salaire net moyen mensuel en EQTP"]*12-df_consolidee['loyer_app_total']*12
df_consolidee["RAV_niv_vie_median_maison"] = df_consolidee["Niveau de vie annuel médian"]*2-df_consolidee["loyer_maison_total"]*12
df_consolidee["RAV_salaire_moyen_eqtp_maison"] = df_consolidee["Salaire net moyen mensuel en EQTP"]*12*2-df_consolidee['loyer_maison_total']*12
df_consolidee["Salaire fixe"] = 2400.0
df_consolidee["RAV_salaire_fixe_app"] = df_consolidee["Salaire fixe"]*12-df_consolidee['loyer_app_total']*12
df_consolidee["RAV_salaire_fixe_maison"] = df_consolidee["Salaire fixe"]*12*2-df_consolidee['loyer_maison_total']*12
df_consolidee
```

```python
import geopandas as gpd
import geoviews as gv
from geoviews import dim

gv.extension('bokeh')
# importing dep geometry
sf = gpd.read_file('../data/departements-version-simplifiee.geojson')
sf.set_index("code", inplace=True)
df_finale = sf.join(df_consolidee)
```

```python
carte1=gv.Polygons(df_finale, vdims=['nom','RAV_niv_vie_median_app','R2_app'])
carte1.opts(width=800, height=800, toolbar='above', color=dim('RAV_niv_vie_median_app'), colorbar=True, tools=['hover'], aspect='equal')
```
